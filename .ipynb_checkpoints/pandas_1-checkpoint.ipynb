```python
import pandas as pd
temp_df = pd.DataFrame({
    'city' : ['bvn', 'rjk', 'gnagar'],
    'temp' : [32,45,46]
})
temp_df
```

```python
wind_df = pd.DataFrame({
    'city' : ['bvn', 'rjk', 'gnagar'],
    'wind' : [9,7,8]
})
wind_df
```

```python
df1 = pd.merge(temp_df, wind_df, on='city')
df1
```

```python
temp_df = pd.DataFrame({
    'city' : ['bvn', 'rjk', 'gnagar','ahemdabad'],
    'temp' : [32,45,46,44]
})
temp_df
```

```python
wind_df = pd.DataFrame({
    'city' : ['bvn', 'rjk', 'jamnagar'],
    'wind' : [9,7,8]
})
wind_df
```

```python
df1 = pd.merge(temp_df, wind_df, on='city')
df1
```

```python
df1 = pd.merge(temp_df, wind_df, on='city', how='outer')
df1
```

```python
df1 = pd.merge(temp_df, wind_df, on='city', how='left')  #left= temp_df
df1
```

```python
df1 = pd.merge(temp_df, wind_df, on='city', how='right')  #right = wind_df
df1
```

```python
df1 = pd.merge(temp_df, wind_df, on='city', how='left', indicator=True)
df1
```
