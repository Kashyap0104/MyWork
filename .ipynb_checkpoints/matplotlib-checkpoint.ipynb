```python
days = [1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t = [43,42,48,46,33,35,37]
avg_t = [45,48,48,46,40,42,41]
```

```python
plt.plot(days,max_t,label='Max')
plt.plot(days,min_t,label='Min')
plt.plot(days,avg_t,label='Avg')
plt.xlabel('Days')
plt.ylabel("Temp")
plt.title('Weather')
# plt.legend(loc='upper right')

plt.legend(loc='best', shadow=True)#if you want to see in upper right corner
plt.grid(color='grey')
```

```python
company=['GOOGLE','AMZ','TESLA','TATA']
revenue=[90,115,98,88]
profit=[40,33,45,55]
```

```python
import numpy as np
xpos=np.arange(len(company))
xpos    #converting to array
```

```python
plt.xticks(ypos,company) #this will assign 1=google ,2=amz etc...
plt.xlabel('Companies')
plt.ylabel('Revenue in US $')
plt.title('US Tech')
plt.bar(xpos,revenue,label='Revenue')
plt.bar(xpos,profit,label='Profit')
plt.legend()
```

```python
plt.xticks(ypos,company) #this will assign 1=google ,2=amz etc...
plt.xlabel('Companies')
plt.ylabel('Revenue in US $')
plt.title('US Tech')
plt.bar(xpos-0.2,revenue,width=0.4,label='Revenue')
plt.bar(xpos+0.2,profit,width=0.4,label='Profit')
plt.legend()
```

```python
plt.yticks(ypos,company) #this will assign 1=google ,2=amz etc...

plt.title('US Tech')
plt.barh(xpos-0.2,revenue,label='Revenue')
plt.barh(xpos+0.2,profit,label='Profit')
plt.legend()  #horizontal
```

```python
blood_sugar_men=[113,86,87,89,90,136,130,123,114,112]
blood_sugar_women=[124,90,100,122,112,89,90,99,110,120]
plt.xlabel('Sugar Range')
plt.ylabel('Total No.of Patient')
plt.hist([blood_sugar_men,blood_sugar_women],bins=[80,100,125,200],rwidth=0.95,color=['green','purple'],
         label=['men','women'])
plt.legend()
```

```python
exp_val = [1800,1600,1200,1000]
exp_labels = ['Home Rent','Food',"Bills",'Other Utilities']
```

```python
plt.pie(exp_val,labels=exp_labels)
```

```python
plt.axis('equal')   #perfect circle
plt.pie(exp_val,labels=exp_labels,radius=1.5)
plt.show()  #don't show labels(which are in output)
```

```python
plt.axis('equal')   #perfect circle
plt.pie(exp_val,labels=exp_labels,radius=1.5,autopct='%0.2f%%')  
#here %0.2f%% the format,where o.2 is 
# indicate in pie chart we want to show  is2 numbers after point(.)
plt.show()
```

```python
plt.axis('equal')   #perfect circle
plt.pie(exp_val,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0,0,0.3,0,])  #explode do this  
plt.show()
```

```python
plt.axis('equal')   #perfect circle
plt.pie(exp_val,labels=exp_labels,radius=1.5,autopct='%0.2f%%',explode=[0,0,0.3,0,],startangle=60)  
#startangle start the chart at angle which will be provide(Home Rent)
plt.show()
```
