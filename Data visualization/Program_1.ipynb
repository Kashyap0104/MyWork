```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
print('Setup Complete')
```

```python
insurance_filepath = 'C:/Users/Dell/Downloads/insurance.csv'
```

```python
insurance_data = pd.read_csv(insurance_filepath)
```

```python
print(insurance_data)
```

```python
insurance_data.head()
```

```python
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['expenses'])
```

```python
sns.regplot(x=insurance_data['bmi'], y=insurance_data['expenses'])
```

```python
#color coded scatter plots
sns.scatterplot(x=insurance_data['bmi'], y=insurance_data['expenses'], hue=insurance_data['smoker'])
```

```python
# sns.lmplot command to add two regression lines, corresponding to smokers and nonsmokers.

sns.lmplot(x='bmi', y='expenses', hue='smoker', data=insurance_data)
```

```python
# Finally, there's one more plot that you'll learn about, that might look slightly 
# different from how you're used to seeing scatter plots. Usually, 
# we use scatter plots to highlight the relationship between two continuous variables (like "bmi" and "charges"). 
# However, we can adapt the design of the scatter plot to feature a categorical variable (like "smoker") on one of the main axes. 
# We'll refer to this plot type as a categorical scatter plot, and we build it with the sns.swarmplot command.

sns.swarmplot(x=insurance_data['smoker'], y=insurance_data['expenses'])
```

```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
print('Setup Complete')
```

```python
iris_filepath = 'C:/Users/Dell/Downloads/iris.csv'
```

```python
iris_data = pd.read_csv(iris_filepath, index_col='Id')
```

```python
print(iris_data)
```

```python
# Histogram 
sns.distplot(a=iris_data['PetalLengthCm'], kde=False)
```

```python
# KDE plot 
sns.kdeplot(data=iris_data['PetalLengthCm'], shade=True)
```

```python
# 2D KDE plot
sns.jointplot(x=iris_data['PetalLengthCm'], y=iris_data['SepalWidthCm'], kind="kde")
```
