```python
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
df = pd.DataFrame(digits.data,columns=digits.feature_names)
df.head(10)
```

```python
digits.target_names
```

```python
digits.target
```

```python
df['target'] = digits.target
df.head(10)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=0.3)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='linear',C=50,gamma='auto')
model.fit(X_train,y_train)
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(SVC(kernel='linear',C=50,gamma='auto'),digits.data,digits.target,cv=4)
```

```python
cross_val_score(SVC(kernel='rbf',C=50,gamma='auto'),digits.data,digits.target,cv=4)
```

```python
cross_val_score(SVC(kernel='linear',C=20,gamma='scale'),digits.data,digits.target,cv=4)
```

```python
cross_val_score(SVC(kernel='linear',C=20,gamma='auto'),digits.data,digits.target,cv=4)
```

```python
import numpy as np
```

```python
kernel = ['rbf','linear']
C = [1,10,20]
avg_scores = {}
for kval in kernel:
    for cval in C:
        cv_scores = cross_val_score(SVC(kernel=kval,C=cval,gamma='auto'),digits.data,digits.target,cv=4)
        avg_scores[kval + '_' +str(cval)] = np.average(cv_scores)
        
avg_scores
```

```python
from sklearn.model_selection import GridSearchCV
gs = GridSearchCV(SVC(gamma='auto'),{
    'C' : [1,10,20],
    'kernel' : ['rbf','linear']
},cv=4,return_train_score=False)

gs.fit(digits.data,digits.target)
gs.cv_results_
```

```python
df = pd.DataFrame(gs.cv_results_)
df
```

```python
gs.best_score_
```

```python
gs.best_params_
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
```

```python
model_params = {
    'svm': {
        'model' : SVC(gamma='auto'),
        'params' : {
            'C' : [1,10,20],
            'kernel' : ['rbf','linear']
        }
    },
    'random_forest' : {
        'model' : RandomForestClassifier(),
        'params' : {
            'n_estimators' : [1,5,10]
        }
    },
    'logostic_regression' : {
        'model' : LogisticRegression(solver='liblinear',multi_class='auto'),
        'params' : {
            'C' : [1,5,10]
        }
    }
}
```

```python
scores = []

for model_name,mp in model_params.items():
    clf = GridSearchCV(mp['model'],mp['params'],cv=5,return_train_score=False)
    clf.fit(digits.data,digits.target)
    scores.append({
        'model' : model_name,
        'best_score' : clf.best_score_,
        'best_params' : clf.best_params_,
    })
```

```python
df = pd.DataFrame(scores,columns=['model','best_score','best_params'])
df
```
